```python
# Phân tích thông kê dữ liệu
df_selected = df[['price','projectid', 'region_name', 'rooms', 'size', 'ward_name', 'price_million_per_m2', 'facilities_key', 'investor_id']]
df_selected.describe()
```

```python
#Kiểm tra dạng phân phối chuẩn của dữ liệu:

sns.displot(df['price'], color='green')
```

```python
# Extract features and target
X = df_selected[['price','projectid', 'region_name', 'rooms', 'size', 'ward_name', 'price_million_per_m2', 'facilities_key', 'investor_id']]  # Additional attributes for prediction
y = df_selected['price']  # Target variable
```

```python
# Step 2: Data Scaling
scaler = MinMaxScaler(feature_range=(0, 1))
X_scaled = scaler.fit_transform(X)
y_scaled = scaler.fit_transform(np.array(y).reshape(-1, 1))
```

```python
# Step 3: Train-Test-Validation Split
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y_scaled, test_size=0.2, shuffle=False)
X_train, X_val, y_train, y_val = train_test_split(X_train, y_train, test_size=0.2, shuffle=False)
```

```python
# Step 4: Create Time Series Dataset
def create_dataset(dataset_X, dataset_y, time_step=1):
    data_X, data_y = [], []
    for i in range(len(dataset_X) - time_step):
        a = dataset_X[i:(i + time_step)]
        data_X.append(a)
        data_y.append(dataset_y[i + time_step])
    return np.array(data_X), np.array(data_y)

time_step = 100  # Length of the time series window
X_train, y_train = create_dataset(X_train, y_train, time_step)
X_val, y_val = create_dataset(X_val, y_val, time_step)
X_test, y_test = create_dataset(X_test, y_test, time_step)

# Reshape input to be [samples, time steps, features] for LSTM
X_train = X_train.reshape(X_train.shape[0], X_train.shape[1], X_train.shape[2])
X_val = X_val.reshape(X_val.shape[0], X_val.shape[1], X_val.shape[2])
X_test = X_test.reshape(X_test.shape[0], X_test.shape[1], X_test.shape[2])

```

```python
# Step 5: Define and Train RNN Model
model = Sequential()

model.add(SimpleRNN(units=16, activation='tanh', return_sequences=True, input_shape=(X_train.shape[1], X_train.shape[2])))
model.add(Dropout(0.2))

model.add(SimpleRNN(units=32, activation='tanh', return_sequences=True))
model.add(Dropout(0.2))

model.add(SimpleRNN(units=128, activation='tanh', return_sequences=True))
model.add(Dropout(0.2))

model.add(SimpleRNN(units=512, activation='tanh'))
model.add(Dropout(0.2))

model.add(Dense(units=1))

model.summary()

model.compile(loss='mean_squared_error', optimizer='adam')

history = model.fit(X_train, y_train, validation_data=(X_val, y_val), epochs=100, batch_size=64, verbose=1,
                    callbacks=[History(), EarlyStopping(monitor='val_loss', patience=10)])
```

```python
# Step 6: Evaluate Model
plt.plot(history.history['loss'], label='train')
plt.plot(history.history['val_loss'], label='validation')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()
plt.show()
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
import numpy as np
# Predict prices on the test set
y_pred = model.predict(X_test)



# Calculate evaluation metrics
mae = mean_absolute_error(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mse)
r2 = r2_score(y_test, y_pred)

# Print the evaluation metrics
print("Model Evaluation Metrics:")
print(f"Mean Absolute Error (MAE): {mae:.2f}")
print(f"Mean Squared Error (MSE): {mse:.2f}")
print(f"Root Mean Squared Error (RMSE): {rmse:.2f}")
print(f"R-squared (R²) Score: {r2:.2f}")
```

```python
# Plot actual vs predicted prices
plt.figure(figsize=(10, 6))
plt.scatter(y_test, y_pred, alpha=0.5)
plt.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], 'r--', lw=2)
plt.xlabel('Actual Prices')
plt.ylabel('Predicted Prices')
plt.title('Actual vs Predicted Prices')
plt.show()
```

Lỗi Tuyệt Đối Trung Bình (MAE): 0.15

MAE đo lường mức độ trung bình của các sai số trong một tập hợp dự đoán, không xem xét hướng của chúng. Đây là trung bình cộng của các sự khác biệt tuyệt đối giữa dự đoán và quan sát thực tế trên mẫu kiểm tra, nơi tất cả các sự khác biệt cá nhân có trọng số như nhau. MAE thấp hơn chỉ ra hiệu suất mô hình tốt hơn. Trong trường hợp này, sai số trung bình trong các dự đoán là 0.15, có thể là tương đối thấp tùy thuộc vào quy mô của biến mục tiêu (giá cả).

Lỗi Bình Phương Trung Bình (MSE): 0.03

MSE đo lường trung bình của các bình phương của các sai số - tức là, trung bình cộng của sự khác biệt bình phương giữa các giá trị ước lượng và giá trị thực tế. MSE nhạy cảm hơn với các ngoại lệ so với MAE vì nó bình phương các sai số trước khi tính trung bình chúng. MSE thấp hơn chỉ ra hiệu suất mô hình tốt hơn. Ở đây, MSE là 0.03 cho thấy rằng các dự đoán của mô hình khá gần với các giá trị thực tế, nhưng mức độ của sai số này phụ thuộc vào quy mô của biến mục tiêu.

Lỗi Bình Phương Trung Bình Căn (RMSE): 0.18

RMSE là căn bậc hai của MSE và nó cung cấp một chỉ số sai số có đơn vị giống như biến mục tiêu. Nó cũng phạt nặng hơn các sai số lớn hơn so với các sai số nhỏ hơn. RMSE là 0.18 chỉ ra độ lệch chuẩn của các sai số dự đoán (sai số dự đoán), cho một cảm giác về mức độ xa các dự đoán so với các giá trị thực tế.

Điểm số R-bình phương (R²): -0.12

R² là một chỉ số thống kê biểu diễn tỷ lệ phần trăm của phương sai cho biến phụ thuộc được giải thích bởi các biến độc lập trong một mô hình hồi quy. Điểm số R² là 1 chỉ ra rằng các dự đoán hồi quy phù hợp hoàn hảo với dữ liệu. Điểm số R² là -0.12 khá kém và chỉ ra rằng mô hình không nắm bắt tốt phương sai trong dữ liệu. Thực tế, điểm số R² âm cho thấy mô hình hoạt động tồi tệ hơn so với một đường ngang (trung bình của biến mục tiêu) sẽ làm. Đây là một chỉ báo rõ ràng rằng các dự đoán của mô hình không đáng tin cậy.

Phân tích và Khuyến nghị: Điểm số R² Thấp: Điểm số R² âm là một chỉ báo mạnh mẽ rằng mô hình không phù hợp với dữ liệu tốt. Điều này có thể do nhiều lý do:

Kiến trúc mô hình có thể không phù hợp với dữ liệu. Có thể thiếu các đặc trưng quan trọng từ mô hình. Dữ liệu có thể không có mối quan hệ đủ mạnh với biến mục tiêu để mô hình học hiệu quả.

MAE, MSE, RMSE Hợp lý: Mặc dù MAE, MSE và RMSE tương đối thấp, những giá trị này một mình không cung cấp một bức tranh đầy đủ. Điểm số R² kém có ưu tiên và cho thấy rằng mặc dù các sai số thấp, mô hình không tổng quát hóa tốt.

Các Bước để Cải Thiện Mô Hình: Kỹ thuật Tính năng:

Đảm bảo rằng tất cả các đặc trưng liên quan được bao gồm. Xem xét tạo ra các đặc trưng mới bắt được thông tin bổ sung từ dữ liệu hiện có.

Kiến trúc Mô hình:

Thử nghiệm với các kiến trúc khác nhau như LSTM hoặc GRU, thường mạnh mẽ hơn SimpleRNN cho dữ liệu chuỗi thời gian. Thử sử dụng nhiều hoặc ít lớp và đơn vị hơn trong mỗi lớp.

Điều chỉnh Tham số:

Thử nghiệm với các tham số khác nhau như tốc độ học, kích thước lô và số lần lặp. Sử dụng các kỹ thuật như tìm kiếm lưới hoặc tìm kiếm ngẫu nhiên để tìm tham số tốt nhất.

Kiểm định chéo:(Cross-validation)

Sử dụng kiểm định chéo k-lần để đảm bảo rằng mô hình không quá khớp và hoạt động tốt trên dữ liệu chưa thấy.

Tiền xử lý Dữ liệu:

Kiểm tra xem có rò rỉ dữ liệu nào không. Đảm bảo rằng dữ liệu được chuẩn hóa/mở rộng đúng cách.

Tăng Kích Thước Dữ liệu:

Thực hiện các bước này có thể giúp cải thiện hiệu suất của mô hình và cung cấp các dự đoán đáng tin cậy hơn.
